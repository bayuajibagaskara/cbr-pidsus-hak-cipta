```python
import os
import json
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Inisialisasi path dokumen
QUERIES_PATH = "data/eval/queries.json"
PREDICTIONS_PATH = "data/results/predictions.csv"
EVAL_DIR = "data/eval"

# Load ground-truth asli dan hasil prediksi sistem CBR
with open(QUERIES_PATH, 'r', encoding='utf-8') as f:
    queries_data = json.load(f)
df_pred = pd.read_csv(PREDICTIONS_PATH)

# Ambil list ID untuk perbandingan
y_true = [q['ground_truth_case_id'] for q in queries_data]
# Mengambil sampel tebakan ID teratas dari string array top_5_case_ids
y_pred = [json.loads(row['top_5_case_ids'])[0] for idx, row in df_pred.iterrows()]

# Kalkulasi metrik evaluasi komputasi hukum menggunakan sklearn
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)

# Simpan laporan hasil evaluasi retrieval ke dalam bentuk CSV terstruktur
eval_summary = [{
    "Metrik": "Retrieval Performance",
    "Accuracy": round(accuracy, 4),
    "Precision": round(precision, 4),
    "Recall": round(recall, 4),
    "F1-Score": round(f1, 4)
}]

df_eval = pd.DataFrame(eval_summary)
output_eval_path = os.path.join(EVAL_DIR, "retrieval_metrics.csv")
df_eval.to_csv(output_eval_path, index=False, encoding='utf-8')

print(f"🎉 Sukses! Hasil kalkulasi performa CBR berhasil disimpan di: {output_eval_path}")
df_eval
```
