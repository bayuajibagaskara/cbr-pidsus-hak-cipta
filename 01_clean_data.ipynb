```python
import os
import re

# Definisikan folder input mentah dan folder output teks bersih
SOURCE_DIR = "data_source"
RAW_DIR = "data/raw"

# Membuat folder otomatis jika belum ada
os.makedirs(RAW_DIR, exist_ok=True)
print("✅ Folder data_source dan data/raw siap digunakan!")
```

```python
def clean_text(text):
    # 1. Mengubah teks menjadi lowercase (huruf kecil semua)
    text = text.lower()
    
    # 2. Menghapus noise teks (Header, Footer, dan Watermark Mahkamah Agung)
    text = re.sub(r"halaman \d+ dari \d+ halaman.*|disclaimer.*|direktori putusan.*|mahkamah agung.*", "", text)
    
    # 3. Normalisasi karakter (hanya menyisakan huruf, angka, dan tanda baca penting)
    text = re.sub(r"[^\w\s\d\.\/\:\-]", " ", text)
    
    # 4. Merapikan spasi yang terlalu renggang
    text = re.sub(r"\s+", " ", text).strip()
    
    return text

print("✅ Fungsi clean_text() berhasil dibuat.")
```

```python
# Cek apakah sudah ada file di folder data_source
if not os.path.exists(SOURCE_DIR) or len(os.listdir(SOURCE_DIR)) == 0:
    print(f"❌ Folder '{SOURCE_DIR}' masih kosong. Silakan masukkan file mentah_001.txt dulu!")
else:
    files = sorted([f for f in os.listdir(SOURCE_DIR) if f.endswith('.txt')])
    
    for idx, filename in enumerate(files, start=1):
        source_path = os.path.join(SOURCE_DIR, filename)
        
        # Membaca file mentah dengan encoding utf-8
        with open(source_path, 'r', encoding='utf-8', errors='ignore') as f:
            isi_mentah = f.read()
        
        # Jalankan proses pembersihan
        teks_bersih = clean_text(isi_mentah)
        
        # Tentukan nama file output sesuai format: case_001.txt
        output_filename = f"case_{str(idx).zfill(3)}.txt"
        output_path = os.path.join(RAW_DIR, output_filename)
        
        # Simpan teks yang sudah bersih
        with open(output_path, 'w', encoding='utf-8') as f:
            f.write(teks_bersih)
            
        print(f"🔹 Sukses membersihkan: {filename} -> {output_filename}")
        
    print(f"\n🎉 Selesai! Semua dokumen bersih tersimpan di: {RAW_DIR}")
```
